```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer


# 1. SET UP THE ENVIRONMENT AND LOAD DATA


# Defining which columns are numbers and which are categories.
# This helps the preprocessing pipeline know what to do with each feature.
numerical_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
categorical_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
target_col = 'target' # This is what we are trying to predict!

# Grabbing the main dataset. Fingers crossed the file path is correct!
df = pd.read_csv('heart_disease_data.csv')

print("Data loaded successfully! Time to get it ready for modeling.")

# Separating our predictors (X) from our desired outcome (y)
X = df.drop(columns=[target_col])
y = df[target_col]


# 2. DATA CLEANING AND FEATURE ENGINEERING


# --- 2.1 Building the Preprocessing Engine (ColumnTransformer) ---

# We need to scale numbers so no single feature (like high cholesterol) dominates the model.
numeric_transformer = StandardScaler()

# We need to one-hot encode categories because models can't handle raw text or numbered categories.
# 'drop="first"' helps us avoid multi-collinearity issues.
categorical_transformer = OneHotEncoder(handle_unknown='ignore', sparse_output=False, drop='first')

# Combining both transformers into a single, efficient pipeline!
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ],
    remainder='passthrough' # Making sure we don't accidentally drop anything else (we shouldn't)
)

# --- 2.2 Running the Engine and Reassembling the DataFrame ---

# Applying all the scaling and encoding magic to our features (X).
X_processed = preprocessor.fit_transform(X)

# Generating the new, expanded list of column names for the transformed data.
feature_names_out = (
    numerical_cols +
    list(preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols))
)

# Putting the clean NumPy array back into a readable Pandas DataFrame.
X_df = pd.DataFrame(X_processed, columns=feature_names_out)

print("\n--- Processed Feature Matrix (X_df) Head ---")
print(X_df.head())
print(f"\nOriginal count of features: {len(X.columns)}")
print(f"New count of features after one-hot encoding: {len(X_df.columns)}")


# 3. FINAL STEP: SPLITTING DATA FOR TRAINING AND TESTING


# Splitting our clean data (80% for training, 20% for testing).
# Setting 'random_state' ensures our split is reproducible!
# Crucially, 'stratify=y' ensures the proportion of 'Disease' cases is the same in both sets.
X_train, X_test, y_train, y_test = train_test_split(
    X_df, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print("\n--- Final Data Split Status ---")
print("Ready for modeling! We've successfully isolated our unseen test set.")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
# Import necessary libraries for modeling and evaluation
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
import time # Need this to see how long our fancy models take to run!

# A nice spot to keep all our results
results = {}

# Defining the models we're going to test out. LR is the crucial baseline!
models = {
    'Logistic Regression': LogisticRegression(random_state=42, solver='liblinear'),
    'Random Forest': RandomForestClassifier(random_state=42),
    'XGBoost': XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
}

print("--- Starting Model Training ---")

# loop through each model, train it, and see how it performs on the test set.
for name, model in models.items():
    start_time = time.time()

    # 1. Teaching the model on the training data
    model.fit(X_train, y_train)

    # 2. Making predictions on the unseen test data
    y_pred = model.predict(X_test)
    # Getting the probability for the AUC-ROC score
    y_proba = model.predict_proba(X_test)[:, 1]

    # 3. Calculate all the key evaluation metrics
    metrics = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred), # SUPER important: how many actual sick people did we catch?
        'F1-Score': f1_score(y_test, y_pred),
        'AUC-ROC': roc_auc_score(y_test, y_proba),
        'Time (s)': time.time() - start_time
    }

    results[name] = metrics

    print(f"\n{name} Trained and Evaluated.")

# Throwing the results into a clean table for easy comparison
results_df = pd.DataFrame(results).T

print("\n--- Model Performance Summary ---")
# Sorting by AUC-ROC to find the current best performer!
print(results_df.sort_values(by='AUC-ROC', ascending=False))

# Identify the winner before we start hyperparameter tuning
best_model_name = results_df['AUC-ROC'].idxmax()
print(f"\nRecommendation: Looks like {best_model_name} is currently the top dog. We'll fine-tune this one next!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import numpy as np

# --- 1. Calculate the Confusion Matrix ---


cm = confusion_matrix(y_test, y_pred)

# Unpack the matrix components for easy interpretation
# Order is: True Negatives, False Positives, False Negatives, True Positives
tn, fp, fn, tp = cm.ravel()
print(f"Confusion Matrix Counts: TN={tn}, FP={fp}, FN={fn}, TP={tp}")

# 2. Plot the Confusion Matrix
plt.figure(figsize=(7, 6))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['No Disease (0)', 'Disease (1)'],
    yticklabels=['No Disease (0)', 'Disease (1)']
)
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix for Untuned Logistic Regression')
plt.show()

# 3. Interpretation of the Matrix
print("\n--- CLINICAL INTERPRETATION ---")
print(f"Total True Cases of Disease (Actual 1s): {fn + tp}")
print(f"Total True Cases of No Disease (Actual 0s): {tn + fp}")
print("-" * 40)
print(f"True Positives (TP = {tp}): Correctly predicted 'Disease'. These patients get needed treatment.")
print(f"True Negatives (TN = {tn}): Correctly predicted 'No Disease'. These patients are correctly reassured.")
print("-" * 40)
print(f"False Negatives (FN = {fn}): The model missed the diagnosis (Predicted 0, Actual 1). **This is the highest clinical risk.**")
print(f"False Positives (FP = {fp}): The model raised a false alarm (Predicted 1, Actual 0). Leads to unnecessary stress/cost.")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, f1_score, make_scorer
import pandas as pd
import time
from sklearn.metrics import accuracy_score, precision_score, recall_score

# --- 1. Define the Scorer ---
# We're prioritizing F1-Score: it finds the best balance between catching the sick (Recall)
# and minimizing false alarms (Precision).
f1_scorer = make_scorer(f1_score)

# --- 2. Initialize the Model ---
# Starting with our basic LR model, ensuring we can reproduce the results later.
lr_model = LogisticRegression(random_state=42)

# --- 3. Define the Parameter Grid ---
# Setting up the grid to test various regularization and solver options.
# We structured this to avoid incompatible combinations that cause errors!
param_grid = [
    # Group 1: Testing L1 (for potential feature selection) and L2 penalties with liblinear.
    {'solver': ['liblinear'],
     'penalty': ['l1', 'l2'],
     'C': [0.001, 0.01, 0.1, 1, 10, 100]},

    # Group 2: Testing L2 penalty with the lbfgs solver, as it's generally fast and robust.
    {'solver': ['lbfgs'],
     'penalty': ['l2'],
     'C': [0.001, 0.01, 0.1, 1, 10, 100]}
]

print("--- Starting Hyperparameter Tuning (Our Goal: Maximize F1-Score Balance) ---")
start_time = time.time()

# --- 4. Initialize and Run GridSearchCV ---
# GridSearchCV will automatically test every single combination defined above.
grid_search = GridSearchCV(
    estimator=lr_model,
    param_grid=param_grid,
    scoring=f1_scorer,  # We are ranking candidates *only* by F1-Score performance.
    cv=5,               # Using 5-fold cross-validation for reliable results.
    n_jobs=-1,          # Using all available processor cores to make this faster.
    verbose=2
)

# Time to start the grid search on our training data.
grid_search.fit(X_train, y_train)
end_time = time.time()

print("\n--- Tuning Complete ---")
print(f"Total Tuning Time: {end_time - start_time:.2f} seconds")

# --- 5. Report Best Results ---
best_params = grid_search.best_params_
best_score = grid_search.best_score_ # This is the highest F1-Score achieved during cross-validation.

print("\nBest Hyperparameters Found (Optimized for F1-Score):")
print(best_params)
print(f"\nCross-Validation F1-Score with Best Params: {best_score:.4f}")

# ----------------------------------------------------------------------
# 6. FINAL MODEL EVALUATION (CALCULATING ALL METRICS)
# ----------------------------------------------------------------------

# The final_tuned_model is the instance that achieved the best F1-Score.
final_tuned_model = grid_search.best_estimator_
y_pred_tuned = final_tuned_model.predict(X_test)
y_proba_tuned = final_tuned_model.predict_proba(X_test)[:, 1]

# Calculating all the final metrics to put in our report.
final_metrics_tuned = {
    'Accuracy': accuracy_score(y_test, y_pred_tuned),
    'Precision': precision_score(y_test, y_pred_tuned),
    'Recall (Sensitivity)': recall_score(y_test, y_pred_tuned),
    'F1-Score': f1_score(y_test, y_pred_tuned),
    'AUC-ROC': roc_auc_score(y_test, y_proba_tuned)
}

print("\nFinal, Optimized Performance on UNSEEN Test Set (All Metrics):")
print(pd.DataFrame(final_metrics_tuned, index=['Optimized LR']).T.round(4))
```

```python
# -----------------------------------------------------------------------
# 1. GENERATE FINAL PERFORMANCE METRICS (ON UNSEEN TEST SET)
# -----------------------------------------------------------------------

print("--- 1. Final Performance Metrics (Tuned LR Model) ---")

# Predict class labels (0 or 1)
y_pred_tuned = final_tuned_model.predict(X_test)
# Predict probabilities for AUC-ROC
y_proba_tuned = final_tuned_model.predict_proba(X_test)[:, 1]

# Calculate all required classification metrics
final_metrics = {
    'Accuracy': accuracy_score(y_test, y_pred_tuned),
    'Precision': precision_score(y_test, y_pred_tuned),
    'Recall (Sensitivity)': recall_score(y_test, y_pred_tuned),
    'F1-Score': f1_score(y_test, y_pred_tuned),
    'AUC-ROC': roc_auc_score(y_test, y_proba_tuned)
}

# Print metrics in a clean format
metrics_df = pd.DataFrame(final_metrics, index=['Tuned Logistic Regression']).T
print(metrics_df)

# -----------------------------------------------------------------------
# 2. FEATURE IMPORTANCE: INTERPRETING COEFFICIENTS AS ODDS RATIOS (ORs)
# -----------------------------------------------------------------------

print("\n--- 2. Feature Importance: Odds Ratios (ORs) ---")

# 1. Get coefficients from the final tuned LR model
coefficients = final_tuned_model.coef_[0]

# 2. Create a DataFrame for interpretation
# FIX: Use X_train.columns instead of X_df.columns to match the model's input size.
feature_importance = pd.DataFrame({
    'Feature': X_train.columns,
    'Coefficient (Beta)': coefficients
})

# 3. Calculate the Odds Ratio (OR = e^Beta)
# OR > 1 means the feature increases the odds of heart disease (positive predictor)
# OR < 1 means the feature decreases the odds of heart disease (negative predictor)
feature_importance['Odds Ratio (e^Beta)'] = np.exp(feature_importance['Coefficient (Beta)'])

# Sort by the absolute magnitude of the coefficient to see the strongest influences
feature_importance['Abs_Coefficient'] = np.abs(feature_importance['Coefficient (Beta)'])
feature_importance = feature_importance.sort_values(by='Abs_Coefficient', ascending=False).drop(columns=['Abs_Coefficient'])

# Print the top 15 most influential features
print(feature_importance.head(15).round(4))

# -----------------------------------------------------------------------
# 3. CONFUSION MATRIX VISUALIZATION
# -----------------------------------------------------------------------

print("\n--- 3. Confusion Matrix Visualization ---")

cm = confusion_matrix(y_test, y_pred_tuned)
plt.figure(figsize=(7, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Disease (0)', 'Disease (1)'],
            yticklabels=['No Disease (0)', 'Disease (1)'])
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix for Tuned Logistic Regression')
plt.show()

# Interpretation Reminder:
print("\nConfusion Matrix Interpretation:")
print(f"True Positives (Correctly diagnosed disease): {cm[1, 1]}")
print(f"False Negatives (Missed diagnoses - most dangerous): {cm[1, 0]}")
print(f"True Negatives (Correctly ruled out disease): {cm[0, 0]}")
print(f"False Positives (False alarms): {cm[0, 1]}")
```
